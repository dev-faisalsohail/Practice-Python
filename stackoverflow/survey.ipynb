```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv('so_survey_subset.csv')

# Print shape and remove missing values
print(df.shape)
df1 = df.dropna()
print(df1.info())

# Extract the column
languages = df1['LanguageWorkedWith']

# Count language occurrences
myctr = Counter()
for lang in languages:
    myctr.update(lang.split(';'))

# Print the counter
print(myctr)

# Convert to DataFrame for plotting
lang_df = pd.DataFrame(myctr.items(), columns=['Language', 'Count'])
lang_df = lang_df.sort_values(by='Count', ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(lang_df['Language'], lang_df['Count'])
plt.title('Most Used Programming Languages')
plt.xlabel('Languages')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('housesdata.csv')
print(df.head())
```

```python
# reading values from dataframe in x and y

x = df['x']
y= df['y']

plt.scatter(x,y)
plt.show()
```

```python
sizes  = df['price'].astype(int)

colors = sizes



plt.scatter(x,y, s=sizes, c=colors, alpha=0.5, edgecolor='black', lw=2)

cbr= plt.colorbar()
cbr.set_label('Price of houses')
plt.xlabel("Distance to plax X")
plt.ylabel("Distance to plax Y")
plt.title("Price of houses in different location")


# saving the figure

plt.savefig('scatter_chart.png') 
plt.show()
```
